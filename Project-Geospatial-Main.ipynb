```python
from pymongo import MongoClient
import pandas as pd

client = MongoClient('mongodb://localhost:27017/')
db = client.companies
```

```python
with_office = db.companies.find({
    "offices":{
        "$not":{
            "$size":0
        }
    }
},{"name":1, "offices":1, "category_code":1, "founded_year":1, "number_of_employees":1})
```

```python
df = pd.DataFrame(with_office)
df.head()
```

```python
#print(df.columns)
#df1=df[['_id','name','category_code','founded_year','number_of_employees','offices']]
#df1.head()
```

```python
def getFirst(data):
    data = data['offices']
#    return (len(data),data[0]['latitude'],data[0]['longitude'])

    # Only create the geoJSON object if all geodata is available
    principal = None
    if data[0]['latitude'] and data[0]['longitude']:
        principal = {
            "type":"Point",
            "coordinates":[data[0]['longitude'], data[0]['latitude']]
        }

    return {
        "totalOffices": len(data),
        "lat": data[0]['latitude'],
        "lng": data[0]['longitude'],
        "headquarters": principal
    }


main_office = df[["offices"]].apply(getFirst, result_type="expand", axis=1)
```

```python

df_geo = pd.concat([df,main_office], axis=1)[["name","lat","lng", "headquarters","number_of_employees", "totalOffices", "category_code","founded_year"]]
df_geo
```
